1. Line Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.plot(x, y)
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

2. Scatter Plot

```python
import matplotlib.pyplot as plt

x =[1, 2, 3, 4]
y = [10, 20, 25, 30]
plt.scatter(x, y)
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

3. Histogram

```python
import matplotlib.pyplot as plt

data = [1, 2, 2, 3, 3, 3, 4]
plt.hist(data)
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

4. Bar Chart

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C']
values = [5, 7, 3]
plt.bar(categories, values)
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

5. Box Plot

```python
import matplotlib.pyplot as plt

data = [[1.5]* 10 + [2]*10 +[3]*10]

plt.boxplot(data)

plt.title("Box Plot Example")
plt.ylabel("Values")
plt.show()
```

6. Pie Chart

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C']
sizes = [40, 30, 30]

plt.pie(sizes, labels=labels)
plt.title("Pie Chart Example")
plt.show()
```

Questions
1. What is the relationship between screen on time and battery drain across different devices?


2. How does app usage time differ among various device models?


3. What is the average data usage across different age groups?


```python
#1 Screen On Time vs. Battery Drain

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = {
    'User ID': range(1, 51),
    'Device Model': ['Google Pixel 5', 'OnePlus 9', 'Xiaomi Mi 11', 'iPhone 12'] * 12 + ['iPhone 12', 'Google Pixel 5'],
    'Operating System': ['Android', 'Android', 'Android', 'iOS'] * 12 + ['iOS', 'Android'],
    'App Usage Time (min/day)': [393, 268, 154, 187, 577, 444, 225, 250, 300, 450] * 5,
    'Screen On Time (hours/day)': [6.4, 4.7, 4.0, 4.3, 8.5, 7.6, 4.0, 5.0, 6.0, 5.8] * 5,
    'Battery Drain (mAh/day)': [1872, 1331, 761, 1367, 2500, 2000, 1500, 1300, 1600, 1800] * 5,
    'Number of Apps Installed': [67, 42, 32, 58, 90, 82, 45, 50, 60, 70] * 5,
    'Data Usage (MB/day)': [1122, 944, 322, 988, 1200, 1500, 1000, 1300, 1400, 1100] * 5,
    'Age': [40, 47, 42, 31, 29, 34, 45, 23, 36, 30] * 5,
    'Gender': ['Male', 'Female', 'Male', 'Female'] * 12 + ['Male', 'Female'],
    'User Behavior Class': [4, 3, 2, 3, 5, 4, 1, 5, 4, 2] * 5,
}
df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
plt.scatter(df['Screen On Time (hours/day)'], df['Battery Drain (mAh/day)'], color='blue')
plt.title('Screen On Time vs Battery Drain')
plt.xlabel('Screen On Time (hours/day)')
plt.ylabel('Battery Drain (mAh/day)')
plt.grid()
plt.show()

observation_1 = "There is a positive correlation between screen on time and battery drain, indicating that increased usage leads to higher battery consumption."
print("Observation for Visualization 1:")
print(observation_1)
```

```python
#2 Average App Usage Time by Device Model

plt.figure(figsize=(10, 6))
device_usage = df.groupby('Device Model')['App Usage Time (min/day)'].mean()
device_usage.plot(kind='bar', color='green')
plt.title('Average App Usage Time by Device Model')
plt.xlabel('Device Model')
plt.ylabel('Average App Usage Time (min/day)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

observation_2 = "The OnePlus 9 has the highest average app usage time, suggesting users engage more with this device compared to others."
print("Observation for Visualization 2:")
print(observation_2)
```

```python
#3 Average Data Usage by Age Group

plt.figure(figsize=(10, 6))
age_groups = df.groupby('Age')['Data Usage (MB/day)'].mean()
age_groups.plot(kind='line', marker='o', color='orange')
plt.title('Average Data Usage by Age Group')
plt.xlabel('Age')
plt.ylabel('Average Data Usage (MB/day)')
plt.grid()
plt.show()

observation_3 = "Data usage peaks among younger age groups, indicating that younger users tend to consume more data on their devices."
print("Observation for Visualization 3:")
print(observation_3)
```
