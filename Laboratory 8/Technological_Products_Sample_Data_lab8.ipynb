```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('TechnologicalProductsData - Tech Products.csv')
```

```python
#A
plt.figure(figsize=(10, 5), facecolor='lightgrey')
plt.plot(data['Date of Release'], data['Sales'], marker='o')
plt.title('Device Sales Over Time')
plt.xlabel('Date of Release')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
#B
df['Year'] = df['Date of Release'].dt.year
total_sales = df.groupby('Year')['Sales'].sum()

plt.figure(figsize=(10, 5), facecolor='lightgrey')
plt.plot(total_sales.index, total_sales.values, linestyle=':', marker='o', label='Sales', linewidth=3, color='red')
plt.title('Total Sales')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.legend(loc="lower right")
plt.grid()
plt.show()
```

```python
#C
plt.figure(figsize=(10, 5), facecolor='lightgrey')
for brand, group in df.groupby('Brand'):
    plt.plot(group['Date of Release'], group['Sales'], label=brand, marker='o')

plt.xlabel('Date of Release')
plt.ylabel('Sales')
plt.title('Device Sales Over Time by Brand')
plt.legend(title='Brand')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
#D
sales_brand = df.groupby('Brand')['Sales'].sum().sort_values()

plt.figure(figsize=(10, 5), facecolor='lightgrey')
plt.bar(sales_brand.index, sales_brand.values)
plt.xlabel('Brand')
plt.ylabel('Total Sales')
plt.title('Total Sales by Brand')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
#E
plt.figure(figsize=(10, 6), facecolor='lightgrey')
df['Country of Origin'].value_counts().plot(kind='bar', color='pink')
plt.xlabel('Country of Origin')
plt.ylabel('Number of Devices')
plt.title('Distribution of Devices by Country of Origin')
plt.grid()
plt.show()
```

```python
#F
plt.figure(figsize=(8, 8), facecolor='lightgrey')
countrycounts = df["Country of Origin"].value_counts()
plt.pie(countrycounts, labels=countrycounts.index, autopct='%1.1f%%', startangle=90)
plt.title("Distribution of Brands by Country of Origin")

plt.show()
```
