```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('TechnologicalProductsData - Tech Products.csv')
```

```python
# Plot
plt.figure(figsize=(10, 5))
plt.plot(data['Date of Release'], data['Sales'], marker='o', label='Sales', linewidth = 3, color ='red')
plt.title('Device Sales Over Time')
plt.xlabel('Date of Release')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()
```
