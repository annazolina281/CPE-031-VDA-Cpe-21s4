```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
data = pd.read_csv('Heart2022.csv')
```

```python
#Question: How does self-rated general health correlate with the average number of days individuals experience poor physical health?
avg_physical_health_days = data.groupby('GeneralHealth')['PhysicalHealthDays'].mean()

plt.figure(figsize=(10, 5), facecolor='lightgrey')
avg_physical_health_days.plot(kind='bar', color='skyblue', edgecolor='black')


plt.title('Average Physical Health Days by General Health Status')
plt.xlabel('General Health')
plt.ylabel('Average Physical Health Days')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Observation:
# Individuals with poorer self-rated general health ("Poor" or "Fair") experience more days with physical health issues.
#Conversely, those reporting better health ("Excellent" or "Very good") have fewer physical health days, indicating a link between perceived health and physical well-being.
```

```python
#Is there a difference in average BMI between individuals who have had a heart attack and those who have not?
avg_bmi_by_heart_attack = data.groupby('HadHeartAttack')['BMI'].mean()


plt.figure(figsize=(10, 5), facecolor='lightgrey')
avg_bmi_by_heart_attack.plot(kind='bar', color='lightcoral', edgecolor='black')

plt.title('Average BMI by Heart Attack Status')
plt.xlabel('Had Heart Attack')
plt.ylabel('Average BMI')
plt.xticks(rotation=0)
plt.grid(True)
plt.tight_layout()
plt.show()

# Observation:
# Yes, there is a difference. The bar plot analysis reveals that individuals who have had a heart attack generally have higher average BMI
# values compared to those who have not. This difference suggests a possible correlation between higher BMI and an increased risk of heart attacks.
```
