```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.array([0.5, 1, 1.3, 1.6, 2, 2.2, 2.5, 3.1, 3.9, 4.4])
y = np.array([1.9, 1.5, 1.2, 1, 0.8, 0.78, 0.65, 0.46, 0.3, 0.23])

# Calculate the slope and y-intercept of the regression line
slope, intercept = np.polyfit(x, y, 1)

# Calculate the predicted values of y using the regression line equation
y_pred = slope * x + intercept

# Plot the original data points and regression line
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.gca().set_facecolor('white')
plt.grid()

# Calculate the correlation coefficient
correlation_matrix = np.corrcoef(x, y)
correlation = correlation_matrix[0, 1]
plt.title('Scatter Plot and Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"Correlation Coefficient: {correlation}")
print(f"The regression line's downward slope indicates a negative linear relationship between x and y.")
```
