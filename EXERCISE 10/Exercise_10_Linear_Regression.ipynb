```python
import numpy as np
import matplotlib.pyplot as plt

#Generate data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

#Calculate the slope and y-intercept of the regression line
slope, intercept = np.polyfit(x, y, 1)

#Calculate the predicted values of y using the regression line equation
y_pred = slope * x + intercept

#Plot the original data points and regression line
plt.scatter(x, y)
plt.plot(x, y_pred, color='red')
plt.gca().set_facecolor('white')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.array([60, 90, 70, 85, 80, 65, 75, 60, 75, 70])  # Exam Scores
y = np.array([70, 95, 70, 75, 90, 75, 75, 60, 80, 65])  # Course Grades

# Calculate the slope and y-intercept of the regression line
slope, intercept = np.polyfit(x, y, 1)

# Calculate the predicted values of y using the regression line equation
y_pred = slope * x + intercept

# Plot the original data points and regression line
plt.scatter(x, y, color='blue', label='Data Points')  # scatter plot of the data points
plt.plot(x, y_pred, color='red', label='Regression Line')  # regression line
plt.grid(True)
plt.title('Scatter Plot and Regression Line')
plt.xlabel('Exam Scores')
plt.ylabel('Course Grades')
plt.legend()
plt.show()

# Calculate the correlation coefficient
correlation_matrix = np.corrcoef(x, y)
correlation = correlation_matrix[0, 1]

# Display the slope, intercept, and correlation coefficient
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"Correlation Coefficient: {correlation}")
print(f"There is a slight positive correlation between students' exam results and their grades, as indicated by the scatter plot and regression line.")
```
