```python
# 1. Question 1: Does the average number of sleep hours differ across levels of general health?
sleep_health = df.groupby('GeneralHealth')['SleepHours'].mean()

# Plotting a bar chart
plt.figure(figsize=(8, 5))
sleep_health.plot(kind='bar', color='pink')
plt.title('Average Sleep Hours by General Health')
plt.xlabel('General Health')
plt.ylabel('Average Sleep Hours')
plt.show()

# This investigation examines the possible relationship between people's self-reported general health status and the number of hours they sleep.
# According to the hypothesis, people who report "better" general health might have healthier sleep habits and, on average, get more sleep than people who report worse health.

#Observation:
# There is a positive correlation, since better health status is associated with longer sleep duration.
# There is a positive correlation, since poor health status is associated with shorter sleep duration.
```

```python
# Question 2: Is there a relationship between alcohol consumption and heart attack incidence?
# Cross-tabulation between AlcoholDrinkers and HadHeartAttack
heart_attack_alcohol = pd.crosstab(df['AlcoholDrinkers'], df['HadHeartAttack'])

# Plotting a stacked bar chart
heart_attack_alcohol.plot(kind='bar', stacked=True, color=['lightcoral', 'lightgreen'], figsize=(8, 5))
plt.title('Heart Attack Incidence by Alcohol Consumption')
plt.xlabel('Alcohol Consumption')
plt.ylabel('Number of People')
plt.legend(title='Had Heart Attack')
plt.show()

# The main objective of this investigation is to find out if there is a clear correlation between past heart attacks and alcohol use.
# Aims to determine whether the percentage of individuals who have experienced a heart attack varies between alcohol users and non-users.
# The "No" Category represents the non alchohol drinkers, and the "Yes" category for the alchohol drinkers.

# Observation:
# There is a positive correlation that there is a higher likelihood of individuals who consume alcohol having a history of heart attack compared to those who do not drink.
# This means that within the dataset, alcohol consumption and heart attack incidence are positively correlatedas one increases (alcohol drinkers), so does the other (heart attack history).
# But this does not imply that alchohol are the main cause of heart attacks.
```
